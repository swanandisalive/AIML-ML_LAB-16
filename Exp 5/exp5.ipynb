## Expt.-5: Disease Prediction using Naive Bayes and Neural Network with Comparison of Classifiers

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, classification_report
```

```python
# Load dataset 
import pandas as pd


data = load_breast_cancer()

X = data.data
y = data.target

print("Classes:", data.target_names)
# malignant, benign

print(X.shape, y.shape)
df = pd.DataFrame(X)

#df['target'] = y
#print(df.head())
#print(df.tail())
df.describe()

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
# Logistic Regression
lr = LogisticRegression(max_iter=5000)
lr.fit(X_train, y_train)
pred_lr = lr.predict(X_test)
print("Logistic Regression Test Accuracy:", accuracy_score(y_test, pred_lr))
pred_lr_train = lr.predict(X_train)
print("Logistic Regression Train Accuracy:", accuracy_score(y_train, pred_lr_train))
```

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
pred_dt = dt.predict(X_test)
print("Decision Tree Test Accuracy:", accuracy_score(y_test, pred_dt))
pred_dt_train = dt.predict(X_train)
print("Decision Tree Train Accuracy:", accuracy_score(y_train, pred_dt_train))
```

```python
# K-Nearest Neighbors
knn = KNeighborsClassifier()
knn.fit(X_train, y_train)
pred_knn = knn.predict(X_test)
print("KNN Test Accuracy:", accuracy_score(y_test, pred_knn))
pred_knn_train = knn.predict(X_train)
print("KNN Train Accuracy:", accuracy_score(y_train, pred_knn_train))
```

```python
# Naive Bayes Classifier
nbc = GaussianNB()
nbc.fit(X_train, y_train)
pred_nbc = nbc.predict(X_test)
print("Naive Bayes Test Accuracy:", accuracy_score(y_test, pred_nbc))
pred_nbc_train = nbc.predict(X_train)
print("Naive Bayes Train Accuracy:", accuracy_score(y_train, pred_nbc_train))
```

```python
# Neural Network
nn = MLPClassifier(max_iter=5000)
nn.fit(X_train, y_train)
pred_nn = nn.predict(X_test)
print("Neural Network Test Accuracy:", accuracy_score(y_test, pred_nn))
pred_nn_train = nn.predict(X_train)
print("Neural Network Train Accuracy:", accuracy_score(y_train, pred_nn_train))
```

```python
# Confusion Matrix and Type I & II Errors for Logistic Regression

cm_lr = confusion_matrix(y_test, pred_lr)
print("Logistic Regression Confusion Matrix:\n", cm_lr)
TN_lr, FP_lr, FN_lr, TP_lr = cm_lr.ravel()
type1_error_lr = FP_lr / (FP_lr + TN_lr)
type2_error_lr = FN_lr / (TP_lr + FN_lr)
print("Logistic Regression Type 1 Error:", type1_error_lr)
print("Logistic Regression Type 2 Error:", type2_error_lr)
print("Type-I error rate in Logistic Regression in %:", type1_error_lr*100)
print("Type-II error rate in Logistic Regression in %:", type2_error_lr*100)
```

```python
# Confusion Matrix and Type I & II Errors for Decision Tree
cm_dt = confusion_matrix(y_test, pred_dt)
print("Decision Tree Confusion Matrix:\n", cm_dt)
TN_dt, FP_dt, FN_dt, TP_dt = cm_dt.ravel()
type1_error_dt = FP_dt / (FP_dt + TN_dt)
type2_error_dt = FN_dt / (TP_dt + FN_dt)
print("Decision Tree Type 1 Error:", type1_error_dt)
print("Decision Tree Type 2 Error:", type2_error_dt)
print("Type-I error rate in Decision Tree in %:", type1_error_dt*100)
print("Type-II error rate in Decision Tree in %:", type2_error_dt*100)
```

```python
# Confusion Matrix and Type I & II Errors for KNN
cm_knn = confusion_matrix(y_test, pred_knn)
print("KNN Confusion Matrix:\n", cm_knn)
TN_knn, FP_knn, FN_knn, TP_knn = cm_knn.ravel()
type1_error_knn = FP_knn / (FP_knn + TN_knn)
type2_error_knn = FN_knn / (TP_knn + FN_knn)
print("KNN Type 1 Error:", type1_error_knn)
print("KNN Type 2 Error:", type2_error_knn)
print("Type-I error rate in KNN in %:", type1_error_knn*100)
print("Type-II error rate in KNN in %:", type2_error_knn*100)
```

```python
# Confusion Matrix and Type I & II Errors for Naive Bayes Classifier
cm_nbc = confusion_matrix(y_test, pred_nbc)
print("Naive Bayes Confusion Matrix:\n", cm_nbc)
TN_nbc, FP_nbc, FN_nbc, TP_nbc = cm_nbc.ravel()
type1_error_nbc = FP_nbc / (FP_nbc + TN_nbc)
type2_error_nbc = FN_nbc / (TP_nbc + FN_nbc)
print("Naive Bayes Type 1 Error:", type1_error_nbc)
print("Naive Bayes Type 2 Error:", type2_error_nbc)
print("Type-I error rate in Naive Bayes in %:", type1_error_nbc*100)
print("Type-II error rate in Naive Bayes in %:", type2_error_nbc*100)
```

```python
# Confusion Matrix and Type I & II Errors for Neural Network
cm_nn = confusion_matrix(y_test, pred_nn)
print("Neural Network Confusion Matrix:\n", cm_nn)
TN_nn, FP_nn, FN_nn, TP_nn = cm_nn.ravel()
type1_error_nn = FP_nn / (FP_nn + TN_nn)
type2_error_nn = FN_nn / (TP_nn + FN_nn)
print("Neural Network Type 1 Error:", type1_error_nn)
print("Neural Network Type 2 Error:", type2_error_nn)
print("Type-I error rate in Neural Network in %:", type1_error_nn*100)
print("Type-II error rate in Neural Network in %:", type2_error_nn*100)
```

```python
# Other metrics - Precision, Recall, F1 Score, ROC AUC Score for Logistic Regression
precision_score_lr = precision_score(y_test, pred_lr)
recall_score_lr = recall_score(y_test, pred_lr)
f1_score_lr = f1_score(y_test, pred_lr)
roc_auc_score_lr = roc_auc_score(y_test, pred_lr)
print("Logistic Regression Precision:", precision_score_lr)
print("Logistic Regression Recall:", recall_score_lr)
print("Logistic Regression F1 Score:", f1_score_lr)
print("Logistic Regression ROC AUC Score:", roc_auc_score_lr)
print("\n\n")

precision_score_dt = precision_score(y_test, pred_dt)
recall_score_dt = recall_score(y_test, pred_dt) 
f1_score_dt = f1_score(y_test, pred_dt)
roc_auc_score_dt = roc_auc_score(y_test, pred_dt)
print("Decision Tree Precision:", precision_score_dt)
print("Decision Tree Recall:", recall_score_dt)
print("Decision Tree F1 Score:", f1_score_dt)
print("Decision Tree ROC AUC Score:", roc_auc_score_dt)
print("\n\n")


precision_score_knn = precision_score(y_test, pred_knn)
recall_score_knn = recall_score(y_test, pred_knn)
f1_score_knn = f1_score(y_test, pred_knn)
roc_auc_score_knn = roc_auc_score(y_test, pred_knn)
print("KNN Precision:", precision_score_knn)
print("KNN Recall:", recall_score_knn)
print("KNN F1 Score:", f1_score_knn)
print("KNN ROC AUC Score:", roc_auc_score_knn)
print("\n\n")

precision_score_nbc = precision_score(y_test, pred_nbc)
recall_score_nbc = recall_score(y_test, pred_nbc)
f1_score_nbc = f1_score(y_test, pred_nbc)
roc_auc_score_nbc = roc_auc_score(y_test, pred_nbc)
print("Naive Bayes Precision:", precision_score_nbc)
print("Naive Bayes Recall:", recall_score_nbc)
print("Naive Bayes F1 Score:", f1_score_nbc)
print("Naive Bayes ROC AUC Score:", roc_auc_score_nbc)  
print("\n\n")

precision_score_nn = precision_score(y_test, pred_nn)
recall_score_nn = recall_score(y_test, pred_nn)
f1_score_nn = f1_score(y_test, pred_nn)
roc_auc_score_nn = roc_auc_score(y_test, pred_nn)
print("Neural Network Precision:", precision_score_nn)
print("Neural Network Recall:", recall_score_nn)
print("Neural Network F1 Score:", f1_score_nn)
print("Neural Network ROC AUC Score:", roc_auc_score_nn)
```
